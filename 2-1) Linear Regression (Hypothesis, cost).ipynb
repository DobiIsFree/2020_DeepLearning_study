# Predicting exam score: regression


## 1. Regression (data)

> training data

|x|y|
|---|---|
|1|1|
|2|2|
|3|3|

<br/>

(Linear) Hypothesis : 어떤 것이 data에 더 맞는 것일까, 일차방정식의 가설 세우기

> H(x) = Wx + b


## 2. Cost function (=Loss function)

- how fit the line to training data  
H(x) - y

> (H(x) - y)^2  
일정하게 차이를 양수로 표현

> cost = 1/m Sigma[1,m] ( H(x^i) - y^i )^2  (m = data set)

## 3. Goal: Minimize cost

minimize cost(W, b)

학습한다 --> W, b를 조절해서 cost function minimize


***

# (1) Build graph using TF operations

### H(x) = Wx + b

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
# X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# hypothesis Wx + b
hypothesis = x_train * W + b

# cost function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))   # reduce_mean : 평균
```

### Gradient Descent

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)   # ... Learning ...
```



# (2, 3) Run/Update graph and get results

```python
# Launch the graph in a session
sess = tf.Session()
# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())

# Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```


## * Placeholders

만들어진 모델에 대하여 값을 다르게 줄때

```python
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

hypothesis = X * W + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [1, 2, 3, 4, 5],
                                                                        Y: [2.1, 3.1, 4.1, 5.1, 6.1]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```

```python
# Testing model
print(sess.run(hypothesis, feed_dict={X:[5]}))
print(sess.run(hypothesis, feed_dict={X:[2.5]}))
print(sess.run(hypothesis, feed_dict={X:[1.5, 3.5]}))
```
